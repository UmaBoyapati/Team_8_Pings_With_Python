```python
#16.Plot a graph to show the distribution of expected length of stay.
x_axis = np.arange(2, 16)      # Plot between (2,16)
mean = statistics.mean(x_axis) # Calculating mean
sd = statistics.stdev(x_axis) # Calculating standard deviation
plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
title = "Distribution of expected length of stay"
plt.title(title)
plt.show()
```
